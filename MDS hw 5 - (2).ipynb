```python
from numpy import *
from pylab import *
from random import *
```

```python
# initial population 
# xo = 0.3 
# yo = 0.3 + 10**-9 
xo = 0.3
yo = 0.3 + 10**-10
x = []
y = []
r = 3.91
time = []
diff = []

x.append(xo)
y.append(yo)
diff.append(log(abs(xo-yo)))
time.append(0)

for k in range(1,40):
    xo = r*xo*(1-xo)
    yo = r*yo*(1-yo)
    time.append(k)
    x.append(xo)
    y.append(yo)
    diff.append(log(abs(xo-yo)))

coeff = polyfit(time, diff, 1)
print("slope : %f, y_intercept : %f" %(coeff[0],coeff[1]))

lin_reg = []

a,b = coeff
for i in range(len(time)):
    lin_reg.append(a*i+b)

plot(time, diff, 'b-', time, lin_reg, 'r-')
```

### Then, ln(diff) = an+b <-> b = ln(diff(0))이므로, ln(diff/diff(0)) = an
diff/diff(0) = exp(an)
따라서 lambda that we wanted to find is same with a = 0.57
